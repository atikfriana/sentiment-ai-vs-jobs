```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset from the local file
df = pd.read_csv('/content/ai_job_market_insights.csv')
print("This the AI job data")
print(df)

# Print the description of the dataset
print("\nDescription of the dataset:")
print(df.info())

df[numerical_cols] = df[numerical_cols].fillna(method='ffill')

# Sort the dataset by 'Salary' for better visualization
plt.figure(figsize=(20, 10))
plt.bar(df['Job Ttile (Domain)'], df['Salary'], color='blue')
plt.xlabel("Job Title (Domain)")
plt.ylabel("Salary")
plt.title('AI Job Market Insights: Top 20 Salary Bar Chart')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()

# Create a pie chart
plt.pie(df['Job Title'].value_counts(), labels=df['Job Title'].unique(), autopct='%1.1f%%')
plt.title('Distribution of Job Categories')
plt.show()

# Analyze the dataset
print("\nSummary statistics of the dataset:")
print(df.describe())
```
